```python
months=[31,28,31,30,31,30,31,31,30,31,30,31]
N=len(df.index)
Timestamp=np.zeros(N)
for i in range(N):    
    day=int(df['Date'][i][:2])
    month=int(df['Date'][i][3:5])
    year=int(df['Date'][i][6:])
    days=day+sum(months[:month-1])
    if np.mod(year,4)==0:
        tot_days=366.
        if month>2:
            days=days+1
    else:
        tot_days=365.
    Timestamp[i]=year+days/tot_days    
    
day, month, year
df['Time']=Timestamp
df.head()
```

 ## Insights
1. Pound sterling has always the highest exchange value with Indian INR followed by Euro which has its value consistently lower than Pound sterling but follows the same pattern over time.
2. US Dollar had higher value than Euro from 2000 till 2003 when it started declining and could never go higher than Euro.
3. Japanese Yen consistently the lowest value except for the years from 2009 till 2013 when Us Dollar was at recession and Japanese Yen was at its peak.
4. From 2011 onwards, US Dollar shows a steeper uptrend as compared to Japanese Yen.
5. All the currencies have a significant trend effects.

```python
y_norm = np.convolve(np.ones_like(y), w, mode='full')
valid_indices = (y_norm != 0)
y_norm = y_norm[valid_indices]

y_w = np.convolve(y, w, mode='full')[valid_indices] / y_norm

# trick: convolve with x-coordinate to find the center of the window at
#        each point.
x_w = np.convolve(x, w, mode='full')[valid_indices] / y_norm
```
